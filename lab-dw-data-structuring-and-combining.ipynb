## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url_2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'

# Read the CSV file into a DataFrame
customer_data_df = pd.read_csv(url)
display(customer_data_df)

customer_data_df_2 = pd.read_csv(url_2)
display(customer_data_df_2)

customer_data_df_3 = pd.read_csv(url_3)
display(customer_data_df_3)
```

```python
#renaming columns for 3

#renaming columns for 1 & 2

customer_data_df_3.rename(columns = {'Customer': 'customer', 
                         'State' : 'state', 
                         'Gender' : 'gender',
                         'Education' : 'education',
                         'Customer Lifetime Value' : 'customer_lifetime_value',
                         'Income' : 'income',
                         'Monthly Premium Auto' : 'monthly_premium_auto',
                         'Number of Open Complaints' : 'number_of_open_complaints',
                         'Policy Type' : 'policy_type',
                         'Vehicle Class' : 'vehicle_class',
                         'Total Claim Amount' : 'total_claim_amount'}, inplace=True)
```

```python
#merging df 1 and df 2
df_1_2 = pd.concat([customer_data_df, customer_data_df_2], axis=0)
display(df_1_2)

df_1_2['Vehicle Class'].value_counts()
```

```python
#renaming columns for 1 & 2

df_1_2.rename(columns = {'Customer': 'customer', 
                         'ST' : 'state', 
                         'GENDER' : 'gender',
                         'Education' : 'education',
                         'Customer Lifetime Value' : 'customer_lifetime_value',
                         'Income' : 'income',
                         'Monthly Premium Auto' : 'monthly_premium_auto',
                         'Number of Open Complaints' : 'number_of_open_complaints',
                         'Policy Type' : 'policy_type',
                         'Vehicle Class' : 'vehicle_class',
                         'Total Claim Amount' : 'total_claim_amount'}, inplace=True)
df_1_2
```

```python
# assigning only M and F for gender

gender_mapping = {'male': 'M', 'female': 'F', 'Femal' : 'F', 'Male' : 'M', 'Female' : 'F'}
df_1_2['gender'] = df_1_2['gender'].replace(gender_mapping)
df_1_2['gender'].unique()
```

```python
# replacing abbreviations of states 

replacing_abbreviations = {'AZ': 'Arizona', 'Cali': 'California', 'WA' : 'Washington'}
df_1_2['state'] = df_1_2['state'].replace(replacing_abbreviations)
display(df_1_2['state'])
```

```python
#*In education, "Bachelors" could be replaced by "Bachelor"*

rename_education = {'Bachelors': 'Bachelor'}
df_1_2['education'] = df_1_2['education'].replace(rename_education)
display(df_1_2['education'])
```

```python
#*In Customer Lifetime Value, delete the `%` character*

df_1_2['customer_lifetime_value'] = df_1_2['customer_lifetime_value'].str.replace('%', '')
display(df_1_2['customer_lifetime_value'])
```

```python
#Number of open complaints formating 

df_1_2['number_of_open_complaints'] = df_1_2['number_of_open_complaints'].fillna(0).astype(int)
```

```python
# merging df 1, 2 & 3 using outer

df_1_2_3 = pd.concat([df_1_2, customer_data_df_3], axis=0, join="outer")
df_1_2_3
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# analyze which sales channel brought the most sales 
pivot_marketing_df = marketing_df.pivot_table(values='total_claim_amount', index='sales_channel', aggfunc='sum').round(2)

display(pivot_marketing_df)
```


### Analysis and Insights:

2. **Branch Performance**:
   - The **Branch** channel has the second highest total claim amount of **$1,301,204.00**.
   - Branch offices are likely serving a large number of customers, making it a critical channel for the company. This may also indicate that customers prefer in-person service for their claims.

3. **Call Center**:
   - The **Call Center** has a total claim amount of **$926,600.82**.
   - Call centers handle a significant volume of claims, indicating that many customers prefer resolving their claims over the phone, perhaps for convenience or efficiency.

4. **Web Channel**:
   - The **Web** channel has the lowest total claim amount of **$706,600.04**.
   - Although it has the lowest total, it shows that a considerable number of customers are comfortable using online services for their claims. This could represent a growth opportunity, especially as more people become accustomed to digital services.

### Further Insights:

1. **Channel Effectiveness**:
   - Agents are likely highly effective in resolving claims and may have better customer service, leading to a higher volume of processed claims.
   - Branches, while second, might offer a balance between in-person service and volume.
   - Call Centers offer a middle ground, providing convenience without the need for physical presence.
   - Web services, although currently the lowest, could be the most cost-effective and scalable with the right improvements in user experience and trust.

2. **Customer Preferences**:
   - Customers might prefer personal interaction (Agent and Branch) for claims, potentially because of the perceived importance and complexity of the claims process.
   - The lower total for Web claims suggests either a lower adoption rate of the digital channel or lower trust/comfort levels with handling claims online.

3. **Strategic Decisions**:
   - Investing in enhancing the web channel could yield significant benefits, as it is cost-effective and scalable.
   - Training and resource allocation to agents should continue, given their high performance.
   - Branches remain important for those who prefer face-to-face interactions.
   - Call centers might benefit from additional support and technology to handle a growing volume of claims.

### Recommendations:

1. **Enhance Digital Presence**:
   - Improve the web platform to make it more user-friendly and secure to increase its adoption for claim processing.
   
2. **Support High-Performing Channels**:
   - Continue to support and possibly expand the agent network and branch services due to their high claim volumes.

3. **Increase Efficiency**:
